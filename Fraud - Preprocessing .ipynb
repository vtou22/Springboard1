For the first round of Preprocessing and Modeling, I will not drop some of the features that are similiar for fraud and normal transactions

```python
df.info()
```

```python
#Drop customerId and transactionDateTime

df.drop(columns=['customerId', 'transactionDateTime'],inplace=True)
```

```python
#correct some dtypes

df[['posEntryMode','posConditionCode']] = df[['posEntryMode','posConditionCode']].astype('O')

#accountOpenDate and dateOfLastAddressChange should be datetime

df['accountOpenDate']=pd.to_datetime(df['accountOpenDate'])
df['dateOfLastAddressChange']=pd.to_datetime(df['dateOfLastAddressChange'])
```

```python
#Turn dates into numeric by ordinal dates

df['accountOpenDate'] = df['accountOpenDate'].map(lambda x:x.toordinal())

df['dateOfLastAddressChange'] = df['dateOfLastAddressChange'].map(lambda x:x.toordinal())
```

```python
#Look at object columns

list(df.select_dtypes('O').columns)
```

```python
#Get dummy variables for Object features

df = pd.get_dummies(df, columns = ['transactionType',
 'merchantName',
 'merchantCategoryCode',
 'acqCountry',
 'posEntryMode',
 'posConditionCode',
 'day_of_week'])
```
